## Importing neccessary modules

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import RepeatedKFold,cross_val_score
from sklearn.model_selection import GridSearchCV
```

### Importing data

```python
estate=pd.read_csv(r"K:\Desktop\NIIT\tables\DS2_C5_S6_KCHouse_Data_Practice.csv")
```

### Data preprocessing

```python
estate.head()
```

```python
estate.info()
```

```python
estate.isna().sum()
```

```python
estate.corr()
```

```python
feature_corr=estate.corr()[["price"]].sort_values("price",ascending=False)[1:]
```

```python
x=estate.loc[:,feature_corr.index[:4]]# Top 3 correlations
y=estate.price
```

### Train test split and scaling

```python
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x_scaled,y,test_size=0.3,random_state=13)
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)

print("Training score of Model : ",model.score(xtrain,ytrain))
print("Root mean squared error of model :",mse(ypred,ytest,squared=False))
```

### Bagging with RandomForestRegressor 

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
x=estate.loc[:,feature_corr.index]# Top 3 correlations
y=estate.price
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x_scaled,y,test_size=0.3,random_state=13)
```

```python
rf_model=RandomForestRegressor()
rf_model.fit(xtrain,ytrain)
ypred=rf_model.predict(xtest)
print("Training score of Model : ",rf_model.score(xtrain,ytrain))
print("Root mean squared error of model :",mse(ypred,ytest,squared=False))
```

### GridSearchCV 

```python
param_grid = {'n_estimators': [10, 50,75],'max_depth': [None, 5, 10, 20],'min_samples_split': [2, 5, 10]}

grid_search = GridSearchCV(rf_model, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(xtrain, ytrain)


print("Best hyperparameters: ", grid_search.best_params_)


ypred = grid_search.predict(xtest)
print("Root mean squared error of models :", np.sqrt(mse(ypred, ytest)))
```

```python
rf_best=RandomForestRegressor(max_depth=10,min_samples_split=10,n_estimators=50)
rf_best.fit(xtrain,ytrain)
ypred=rf_best.predict(xtest)
```

```python
print("Root Mean Squared error of best model : ",np.sqrt(mse(ypred,ytest)))
```

```python
sns.kdeplot(ypred, fill = True,linewidth=6,label="Predicted")
plt.legend(loc="upper right")
sns.kdeplot(ytest,linestyle="dashdot",fill = True,color="orange",label="Actual")
plt.legend(loc="upper right")
plt.show()
```
