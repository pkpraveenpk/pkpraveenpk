## **Importing neccesary**

```python
import spacy
import pandas as pd
nlp=spacy.load('en_core_web_sm')
import matplotlib.pyplot as plt
import re
```

```python
review1=open(r"/content/Review1.txt").read()
review2=open(r"/content/Review2.txt").read()
review3=open(r"/content/Review3.txt").read()
```

**Task1**

```python
def count_sentiment_words(text):
    docs=nlp(text)
    text=" ".join(str(doc) for doc in docs if(doc.is_stop==False))
    good=['good', 'great', 'excellent', 'positive', 'wonderful', 'amazing', 'fantastic', 'terrific', 'superb', 'outstanding']
    bad=['bad', 'terrible', 'awful', 'horrible', 'negative', 'sad', 'depressing', 'unhappy', 'gloomy', 'bleak']
    neutral=['neutral', 'okay', 'average', 'ordinary', 'typical', 'commonplace', 'mundane', 'routine', 'standard', 'usual']
    good_count=sum([text.lower().count(word) for word in good])
    bad_count=sum([text.lower().count(word) for word in bad])
    neutral_count=sum([text.lower().count(word) for word in neutral])

    return {'good': good_count, 'bad': bad_count, 'neutral': neutral_count}

sentiments1=count_sentiment_words(review1)
print('Review1')
print('The proportion of words in each sentiment : ')
print(sentiments1)
print()
sentiments2=count_sentiment_words(review2)
print('Review2')
print('The proportion of words in each sentiment : ')
print(sentiments2)
print()
sentiments3=count_sentiment_words(review3)
print('Review2')
print('The proportion of words in each sentiment : ')
print(sentiments3)
```

**Task2**

```python
def stop_word_removal(text):
    text=re.sub(r'#.!_*', '',text)
    text=' '.join(text.split())
    nlp=spacy.load("en_core_web_sm")
    stopwords=nlp.Defaults.stop_words
    new_stopwords=['ingredients', 'live', 'person', 'went', 'came', 'gave', 'seen', 'movie']
    stopwords.update(new_stopwords)
    labels=['__label__1', '__label__2', 'label__1', 'label__2']
    tokens=[token for token in nlp(text) if not (token.is_stop or token.tag_ in labels)]
    lemmatized=[token.lemma_ for token in tokens]
    pos_tags=[token.pos_ for token in tokens]
    clean_text=' '.join(str(token) for token in tokens)
    vector=nlp(clean_text).vector
    return clean_text, lemmatized, pos_tags, vector

clean_review1,lemma1,vector1,pos_tag1=stop_word_removal(review1)
clean_review2,lemma2,vector2,pos_tag2=stop_word_removal(review2)
clean_review3,lemma3,vector3,pos_tag3=stop_word_removal(review3)


  
```

**Interpretation:**

Choosing the Review 2 corpus for sentiment analysis is beneficial because it has a balanced distribution of positive, negative, and neutral reviews. This ensures that the model is trained on a representative sample of different types of language it may encounter in the real world. This balanced bias can prevent the model from overfitting to a specific sentiment category, leading to better generalization on new data. Overall, a corpus with a good bias of all sentiments is essential for building an accurate and robust sentiment analysis model.
