```python
txt=open(r"/content/DS3_C2_S4_GettysburgAddress_Data_Challenge.txt").read()
```

**Task2**

```python
def bow_rep(string,dictionary={}):
    word_freq = defaultdict(int)
    string=nlp(string)
    for item in string:
        if (item not in dictionary):
            dictionary[str(item)]=len(dictionary)
        word_freq[dictionary[str(item)]]+=1
    return dictionary,word_freq

vector,bow=bow_rep(txt)
```

```python
print('Vector representation : ',nlp('dedicated').vector)
print()
print('Vector representation : ',nlp('nation').vector)

print()

print('Bag of words representation : ','nation : ',bow[vector['nation']],'dedicated : ',bow[vector['dedicated']])
```

**Task3**

```python
from spacy.matcher import PhraseMatcher

matcher=PhraseMatcher(nlp.vocab)
patterns=[nlp("battlefield"),nlp("battle-field")]
matcher.add("battlefield",patterns)
doc=nlp("battlefield ! battle-field!")
matches=matcher(doc)

for match_id,start,end in matches:
    string_id = nlp.vocab.strings[match_id]
    span = doc[start:end]
    print(span.text)
```
