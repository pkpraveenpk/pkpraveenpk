## Importing neccessary

```python
from scipy.stats import chi2,chi2_contingency,norm
import numpy as np
```

## Task 1

## Ho : The manager's distribution the observed frequencies of responses are consistent with some expected distribution of responses

## H1 : The alternate hypothesis would be that the observed frequencies of responses are not consistent with the expected distribution of responses

```python
observed=[24,108,61,14]
n=len(observed)
exp=sum(observed)/n
```

```python
chi=0
for item in observed:
    chi+=(item-exp)**2/exp
```

```python
print("Chi statistics : ",chi)
```

```python
alpha=0.05
chi_crit=chi2.ppf(1-alpha,df=3)
```

```python
print("Chi critical : ",chi_crit)
```

```python
p_value=norm.sf(chi,11)
```

```python
if p_value < alpha:
    if chi > chi_crit:
        print("Reject the null hypothesis")
    else:
        print("Accept the alternate hypothesis")
else:
    print("Accept the alternate hypothesis")
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,15)
plt.plot(x,chi2.pdf(x,3),color="red")
plt.title("Pdf of chi2")
plt.show()
```

## Task 2

## Ho : The number of bedrooms in a new home is independent of the number of stories. That is, the proportion of homes with three bedrooms is the same for one-story and two-story homes.

## H1 : That the number of bedrooms in a new home is dependent on the number of stories. That is, the proportion of homes with three bedrooms is different for one-story and two-story homes.

```python
alpha=0.1
data=[[116,101,57],[90,325,160]]
```

```python
chi,p_value,df,expected=chi2_contingency(data)

chi_crit=chi2.ppf(1-alpha,df)

print("Chi Critical : ",chi_crit)
```

```python
if p_value < alpha:
    if chi > chi_crit:
        print("Reject the null hypothesis")
    else:
        print("Accept the alternate hypothesis")
else:
    print("Accept the alternate hypothesis")
```

```python
x=np.arange(-1,15)
plt.plot(x,chi2.pdf(x,df),color="red")
plt.title("Pdf of chi2")
plt.show()
```
