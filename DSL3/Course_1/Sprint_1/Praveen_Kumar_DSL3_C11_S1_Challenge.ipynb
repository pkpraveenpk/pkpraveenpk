## Importing neccessary modules

```python
from scipy.stats import chi2,chi2_contingency as cont
import matplotlib.pyplot as plt
```

# Task 1 :


## Ho : The distribution of regular and early decision is the same in 6 universities

## H1 : The distribution of regular and early decision is not the same in 6 universities

```python
data=[[2115,1792,5304,1734,2685,1245],[577,627,1228,444,1195,761]]
```

```python
chi_stats,p_value,dof,expected=cont(data)
```

```python
alpha=0.05
chi_crit=chi2.ppf(1-alpha,dof)
```

```python
if p_value < alpha:
    if chi_stats > chi_crit:
        print("Reject the null hypothesis")
    else:
        print("Accept the null hypothesis")
else:
    print("Accept the null hypothesis")
```

```python
x=[i for i in range(-1,10)]
plt.plot(x,chi2.pdf(x,alpha))
```

## Conclusion:
The distribution of regular and early decision is not the same in 6 universities.

The distribution is not same so the applications must be made more feasible and promoted amongst public to improve the frequency of applications in Early decisions
